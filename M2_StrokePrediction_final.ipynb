## Cleaning Data

```
# Fill in missing values for Smoking Status with "No info"
train_f = df_health.na.fill('No Info', subset=['smoking_status'])
```

```
# Fill in missing values with mean for BMI
from pyspark.sql.functions import mean
mean = train_f.select(mean(train_f['bmi'])).collect()
mean_bmi = mean[0][0]
train_f = train_f.na.fill(mean_bmi,['bmi'])
```

```
from pyspark.ml.feature import (VectorAssembler,OneHotEncoderEstimator,StringIndexer)

# indexing all categorical columns in the dataset
indexer_gender = StringIndexer(inputCol="gender", outputCol="genderIndex")
indexer_ever_married = StringIndexer(inputCol="ever_married", outputCol="ever_marriedIndex")
indexer_work_type = StringIndexer(inputCol="work_type", outputCol="work_typeIndex")
indexer_Residence_type = StringIndexer(inputCol="Residence_type", outputCol="Residence_typeIndex")
indexer_smoking_status = StringIndexer(inputCol="smoking_status", outputCol="smoking_statusIndex")

# perform one hot encoding on indexed data
encoder = OneHotEncoderEstimator(inputCols=["genderIndex","ever_marriedIndex","work_typeIndex","Residence_typeIndex","smoking_statusIndex"],
                                 outputCols=["gender_encoder", "ever_married_encoder","work_type_encoder",
                           "Residence_type_encoder", "smoking_status_encoder"])

#create assembler to combine given list of columns into a single vector column
assembler = VectorAssembler(inputCols=['age','hypertension','heart_disease','avg_glucose_level','bmi','work_type_encoder','Residence_type_encoder','smoking_status_encoder','gender_encoder','ever_married_encoder'],outputCol='features')
# ,
```

```
from pyspark.ml import Pipeline, PipelineModel
from pyspark.ml.feature import Bucketizer, RegexTokenizer, StopWordsRemover, CountVectorizer, IDF, NGram, StringIndexer, VectorAssembler
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier, GBTClassifier, NaiveBayes,DecisionTreeClassifier
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

pipeline = Pipeline(stages=[])

basePipeline =[indexer_gender, indexer_ever_married, indexer_work_type, indexer_Residence_type,
                           indexer_smoking_status, encoder,assembler]
```

```
rf = RandomForestClassifier(labelCol='stroke',featuresCol='features',numTrees=50)
pl_rf = basePipeline + [rf]
pg_rf = ParamGridBuilder()\
      .baseOn({pipeline.stages: pl_rf})\
      .build()

dtc = DecisionTreeClassifier(labelCol='stroke',featuresCol='features')
pl_dtc= basePipeline + [dtc]
pg_dtc = ParamGridBuilder()\
      .baseOn({pipeline.stages: pl_dtc})\
      .build()

lr = LogisticRegression(maxIter=10,labelCol='stroke',featuresCol='features')
pl_lr = basePipeline + [lr]
pg_lr = ParamGridBuilder()\
          .baseOn({pipeline.stages: pl_lr})\
          .addGrid(lr.regParam,[0.01, .08])\
          .addGrid(lr.elasticNetParam,[0.1, 0.8])\
          .build()
    
nb = NaiveBayes(labelCol='stroke',featuresCol='features')
pl_nb = basePipeline + [nb]
pg_nb = ParamGridBuilder()\
.baseOn({pipeline.stages: pl_nb})\
.addGrid(nb.smoothing,[0.8,1.0])\
.build()
```

```
paramGrid =pg_rf+pg_dtc+pg_nb+pg_lr 
```

```
training, testing = train_f.randomSplit([0.7, 0.3], 24)   # proportions [], seed for random
```

```
cv = CrossValidator()\
      .setEstimator(pipeline)\
      .setEvaluator(MulticlassClassificationEvaluator(labelCol="stroke", predictionCol="prediction", metricName="accuracy"))\
      .setEstimatorParamMaps(paramGrid)\
      .setNumFolds(10)
cvModel = cv.fit(training)
```

```
import numpy as np
# BinaryClassificationEvaluator defaults to ROC AUC, so higher is better
# http://gim.unmc.edu/dxtests/roc3.htm
cvModel.getEstimatorParamMaps()[ np.argmax(cvModel.avgMetrics) ]
```

```
cvModel.getEstimatorParamMaps()[ np.argmin(cvModel.avgMetrics) ]
```

###Model Measures

```
import re
def paramGrid_model_name(model):
  params = [v for v in model.values() if type(v) is not list]
  name = [v[-1] for v in model.values() if type(v) is list][0]
  name = re.match(r'([a-zA-Z]*)', str(name)).groups()[0]
  return "{}{}".format(name,params)

# Resulting metric and model description
# get the measure from the CrossValidator, cvModel.avgMetrics
# get the model name & params from the paramGrid
# put them together here:
kmeans_measures = zip(cvModel.avgMetrics, [paramGrid_model_name(m) for m in paramGrid])
metrics,model_names = zip(*kmeans_measures)
```

### Plotting Model Measures

```
import seaborn as sns
import matplotlib.pyplot as plt

plt.clf() # clear figure
fig = plt.figure( figsize=(5, 5))
plt.style.use('fivethirtyeight')
axis = fig.add_axes([0.1, 0.3, 0.8, 0.6])
# plot the metrics as Y
#plt.plot(range(len(model_names)),metrics)
plt.bar(range(len(model_names)),metrics)
# plot the model name & param as X labels
plt.xticks(range(len(model_names)), model_names, rotation=70, fontsize=6)
plt.yticks(fontsize=6)
#plt.xlabel('model',fontsize=8)
plt.ylabel('ROC AUC (better is greater)',fontsize=8)
plt.title('Model evaluations')
display(plt.show())
```

### Running the best model on the test data

```
## Make predictions on test documents. 
# CrossValidator.fit() is in cvModel, which is the best model found (lrModel).
predictions = cvModel.transform(testing)
```
